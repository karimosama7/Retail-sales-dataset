# **What are the types of products and the best sellers?**

```python
pro_data=lap_data['Product Category'].value_counts()
pro_data
```

```python
plt.pie(pro_data.values,labels=pro_data.index,autopct="%1.1f%%")
plt.legend(loc="center right")
plt.show()
```

```python
plt.bar(pro_data.index,pro_data.values,color="g")
plt.title("Product Category count",fontsize=20,color="c")
plt.xlabel("Product Category",fontsize=20,color="c")
plt.ylabel("count",fontsize=20,color="c")
plt.show()
```

```python
sns.barplot(data=lap_data,x=lap_data['Product Category'],y=lap_data['Total Amount'],hue="Gender")
plt.show()
```

```python
sns.barplot(data=lap_data,y=lap_data['Product Category'],x=lap_data['Age'],hue="Gender")
plt.show()
```

```python
lap_data["Quantity"].value_counts()
```

```python
lap_data["Price per Unit"].value_counts()
```

```python
sns.relplot(data=lap_data,y=lap_data['Price per Unit'],x=lap_data['Age'],hue="Product Category",row='Quantity',col="Gender",kind='scatter')
plt.show()
```

```python
lap_data["Date"]=pd.to_datetime(lap_data["Date"])
lap_data["year"]=lap_data["Date"].dt.year
lap_data['months'] = lap_data["Date"].dt.month_name() 
lap_data["year"].value_counts()
```

```python
data_y=lap_data.groupby("year")[["months"]].max().reset_index()
data_y.head()
```

```python
lap_data['Date']=pd.to_datetime(lap_data['Date'])
plt.figure(figsize=(12, 8))
sns.lineplot(data=lap_data, x=lap_data['months'], y=lap_data['Total Amount'], hue=lap_data['Product Category'],marker="o")
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.show()
```

# **The highest number of sales and the highest price**

```python
filtered_data = lap_data[(lap_data["Price per Unit"] == lap_data["Price per Unit"].max()) & (lap_data["Quantity"] == lap_data["Quantity"].max())].reset_index()
filtered_data
```

```python
filtered_data=filtered_data.sort_values(by="Date",ascending=True)
filtered_data
```

```python
filtered_data.info()
```

```python
filtered_data.describe()
```

```python
sns.barplot(data=filtered_data,x=filtered_data['Product Category'],y=filtered_data['Total Amount'],hue="Gender")
plt.show()
```

```python
sns.barplot(data=filtered_data,y=filtered_data['Product Category'],x=filtered_data['Age'],hue="Gender")
plt.show()
```
